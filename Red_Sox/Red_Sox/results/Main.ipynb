# Load files

```python



import pandas as pd
import glob

# Define the path to the CSV files
path = r'C:\Users\clint\Desktop\Coding Task Darthmouth\Red_Sox\Red_Sox'

# Use glob to get all the CSV files in the directory
all_files = glob.glob(path + "/*.csv")

# Combine all the CSV files into a single DataFrame with an additional 'year' column
df_list = []
for file in all_files:
    df_temp = pd.read_csv(file)
    year = file.split('_')[-1].split('.')[0]  # Extract the year from the file name
    df_temp['year'] = year
    df_list.append(df_temp)

df = pd.concat(df_list, ignore_index=True)

df
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binned_statistic

# Assuming 'df' is your DataFrame
x = df['days_from_transaction_until_game']
y = df['price_per_ticket']

# Define the number of bins
n_bins = 30  # Adjust for your data distribution

# Compute binned statistics (mean price per ticket in each bin)
bin_means, bin_edges, _ = binned_statistic(x, y, statistic='mean', bins=n_bins)

# Calculate bin midpoints for plotting
bin_midpoints = (bin_edges[:-1] + bin_edges[1:]) / 2

# Create the plot
plt.figure(figsize=(10, 6))

# Plot raw data as a scatter plot
plt.scatter(x, y, alpha=0.5, label='Raw Data', color='lightgray')

# Overlay the binned scatter plot
plt.scatter(bin_midpoints, bin_means, color='blue', label='Binned Means', zorder=3)
plt.plot(bin_midpoints, bin_means, color='blue', linestyle='--', alpha=0.8, zorder=2)

# Add titles, labels, and grid
plt.title('Days from Transaction Until Game vs Price per Ticket')
plt.xlabel('Days from Transaction Until Game')
plt.ylabel('Price per Ticket')
plt.gca().invert_xaxis()  # Invert the x-axis
plt.grid(True)
plt.legend()
plt.show()
```
